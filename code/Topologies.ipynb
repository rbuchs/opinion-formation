# Topologies
We present the kind of topologies that can be used when initating an `OpinionGraph`. We keep the number of nodes $n=30$ and the number of nodes to number of opinions ratio $\gamma=10$ constant and varry some other parameters.

```python
#Import the relevant modules
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

import OpinionGraph
import OpinionAlgorithm
```

```python
n = 30 # number of nodes
gamma = 10 # ratio #nodes/#opinions
n_opinion = int(n/gamma) #number of opinions
```

## Random graph
A graph is chosen uniformly at random from the set of all graphs with `n` nodes and `m` edges.

```python
for m in [30, 60, 120]: # number of edges
    #Generate random graph
    G1 = OpinionGraph.CreateRandom(n, m, n_opinion)
    #plot the opinion graph
    G1.plot()
```

## BarabsiAlbert
Random graph according to the BarabsiAlbert preferential attachment model.

``A. L. Barabsi and R. Albert Emergence of scaling in random networks, Science 286, pp 509-512, 1999``

```python
for m in [1,2,3]: # Number of edges to attach from a new node to existing nodes
    #Generate Albert Barbasi graph
    G2 = OpinionGraph.CreateBarbasiAlbert(n, m, n_opinion)
    #plot the opinion graph
    G2.plot()
```

```python
?OpinionGraph.CreateBarbasiAlbert
```

## Newman-Watts-Strogatz
NewmanWattsStrogatz small-world graph.

``M. E. J. Newman and D. J. Watts, Renormalization group analysis of the small-world network model, Physics Letters A, 263, 341, 1999. http://dx.doi.org/10.1016/S0375-9601(99)00757-4``

```python
for p in np.arange(0, 1.2, 0.2): # The probability of adding a new edge for each edge.
    print('p =', p)
    for k in [3,5,7]: # Each node is joined with its `k` nearest neighbors in a ring topology.
        #Generate Newman-Watts-Strogatz graph
        G3 = OpinionGraph.CreateNewmanWattsStrogatz(n, k, p, n_opinion)
        #plot the opinion graph
        G3.plot()
        plt.show()
```

```python
?OpinionGraph.CreateNewmanWattsStrogatz
```

## Powerlaw cluster

``P. Holme and B. J. Kim, Growing scale-free networks with tunable clustering, Phys. Rev. E, 65, 026107, 2002.``

```python
for p in np.arange(0, 1.2, 0.2): # Probability of adding a triangle after adding a random edge
    print('p =', p)
    for m in [3,5,7]: # the number of random edges to add for each new node
        #Generate Powerlaw cluster graph
        print('------ m =', m)
        G4 = OpinionGraph.CreatePowerlawCluster(n, m, p, n_opinion, True)
        print('----------- average clustering =', G4.average_clustering())
        #plot the opinion graph
        G4.plot()
        plt.show()
```
